```python
#creating numpy/n-d arrays
import numpy as np
arr1=np.array([1,2,3,4])
# print(arr1)
arr1
```

```python

arr2=np.array([[4,5,6,7],[8,9,2,1]])
arr2
```

```python
arr3=np.zeros((2,3)) #it will create 2 rows and 3 colums and having zeros init. 
arr3
```

```python
arr4=np.ones((3,3))#it will create 3 row and 3 columns and having ones in it.
arr4
```

```python
arr5=np.identity(5) #it will create 5 ones in diagonal
arr5
```

```python
arr6=np.arange(10) #it will create list from 0 to 10
arr6
```

```python
#if you want 5 to 15 and 2 space in it
arr7=np.arange(5,16,2)
arr7
```
